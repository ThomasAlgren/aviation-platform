```python
import pandas as pd
import urllib.request
import zipfile
import os

print("Downloading FAA aircraft registry...")
url = "https://registry.faa.gov/database/ReleasableAircraft.zip"
urllib.request.urlretrieve(url, "faa_registry.zip")

print("Unzipping...")
with zipfile.ZipFile("faa_registry.zip", "r") as z:
    z.extractall("faa_data")

print("Done! Files downloaded:")
for f in os.listdir("faa_data"):
    print(" -", f)
    
```

```python
df = pd.read_csv("faa_data/MASTER.txt", 
                 encoding="latin1", 
                 low_memory=False)

print(f"Total aircraft in FAA registry: {len(df):,}")
print(f"\nColumns available:")
for col in df.columns:
    print(" -", col)
```
