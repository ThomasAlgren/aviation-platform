```python
import os

if os.path.exists("faa_data"):
    print("Found it! Contents:")
    for f in os.listdir("faa_data"):
        print(" -", f)
else:
    print("Not found - need to download")
```

```python
import urllib.request
import zipfile

print("Downloading FAA registry... (this will take a minute)")
urllib.request.urlretrieve("https://registry.faa.gov/database/ReleasableAircraft.zip", "faa_registry.zip")
print("Download done! Unzipping...")
with zipfile.ZipFile("faa_registry.zip", "r") as z:
    z.extractall("faa_data")
print("Done! Files:")
for f in os.listdir("faa_data"):
    print(" -", f)
```
